```python
from typing import Optional

class TreeNode():
    def __init__(self, data:Optional[int]=None):
        self.data = data
        self.left = None
        self.right = None
        
class BinarySearchTree():
    def __init__(self, root:Optional[TreeNode]=None):
        self.root = root
        
    def search(value:int)->Optional[TreeNode]:
        
        node = self.root
        while node and node.data != value:
            node = node.left if value < node.data else node.right
        return node
    
    def insert(self, value:int):
        
        newNode = TreeNode(value)

        if not self.root:
            self.root = newNode
            return
        
        node = self.root
        parentNode = None
        
        while node:
            parentNode = node
            node = node.left if value < node.data else node.right
        
        if value < parentNode.data:
            parentNode.left = newNode
        else:
            parentNode.right = newNode
        
    def delete(self, value:int):
        
        node = self.root
        parentNode = None
        
        while node and node.data != value:
            parentNode = node
            node = node.left if value < node.data else node.right
        
        if not node:
            return
        
        if node.left and node.right:
            smallNodeParent = node.right
            smallNode = node
            while smallNode.left:
                smallNodeParent = smallNode
                smallNode = smallNode.left
            node.data = smallNode.data
            node = smallNode
            parentNode = smallNodeParent
        
        child = node.left if node.left else node.right
        if not parentNode:
            self.root = child
        elif parentNode.left == node:
            parentNode.left = child
        elif parentNode.right == node:
            parentNode.right = child
            
    def in_order(self, node):
        
        if not node:
            return
        
        self.in_order(node.left)
        print(node.data)
        self.in_order(node.right)
        
    def getMax(self):
        
        node = self.root
        while node.right:
            node = node.right
            
        if not node:
            return 0
        else:
            return node.data
        
if __name__ == '__main__':
    bst = BinarySearchTree()
    bst.insert(33)
    bst.insert(16)
    bst.insert(50)
    bst.insert(13)
    bst.insert(18)
    bst.insert(34)
    bst.insert(58)
    bst.insert(15)
    bst.insert(17)
    bst.insert(25)  
    bst.insert(51)
    bst.insert(66)
    bst.insert(19)
    bst.insert(27)
    bst.insert(55)
            
#     bst.in_order(bst.root)
            
    bst.delete(13)
    bst.delete(18)
    bst.delete(55)
    bst.in_order(bst.root)
    print(bst.getMax())
```
