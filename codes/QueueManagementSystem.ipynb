```python
class EQueue():
    size = 0
    items = []
    head = 0
    tail = 0
    
    def __init__(self,size):
        self.size = size
        self.items = list(range(size))
       
    def isEmpty(self):
        return self.head == self.tail
    
    def isFull(self):
        if (self.tail + 1) % self.size == self.head:
            return True
        return False
        
    def pop(self):
        if self.isEmpty():
            return
        
        self.head = (self.head + 1)%self.size
        return self.items[self.head - 1]
    
    def push(self,item):
        if self.isFull():
            print("The queue is full!!!")
            return
        
        self.items[self.tail] = item
        self.tail = (self.tail + 1)%self.size
        
class QueueManagementSystem():
    
    def __init__(self,size):
        self.queue = EQueue(size)
        
    def getServiceNumber(self,item):
        if self.queue.isFull():
            print('The queue is full!!!!')
            return -1
        self.queue.push(item)
        print(f"{item} success to get a service number{self.queue.tail}!")
        return self.queue.tail
    
    def offService(self):
        if self.queue.isEmpty():
            print('there is no one is waiting!!!')
            return
        
        print(f"{self.queue.pop()} is your turn to get service!")
        
        
queueSystem = QueueManagementSystem(10)
queueSystem.getServiceNumber("Li Ming0")
queueSystem.getServiceNumber("Li Ming1")
queueSystem.getServiceNumber("Li Ming2")
queueSystem.getServiceNumber("Li Ming3")
queueSystem.getServiceNumber("Li Ming4")

queueSystem.offService()

queueSystem.getServiceNumber("Han Meimei0")

queueSystem.offService()

queueSystem.getServiceNumber("Han Meimei1")
queueSystem.getServiceNumber("Han Meimei2")
queueSystem.getServiceNumber("Han Meimei3")

queueSystem.offService()

queueSystem.getServiceNumber("Han Meimei4")

queueSystem.getServiceNumber("Han Meimei5")

queueSystem.getServiceNumber("Han Meimei6")

queueSystem.getServiceNumber("Han Meimei7")

queueSystem.getServiceNumber("Han Meimei8")

queueSystem.getServiceNumber("Han Meimei9")

queueSystem.getServiceNumber("Han Meimei10")

queueSystem.getServiceNumber("Han Meimei11")
queueSystem.getServiceNumber("Han Meimei12")


queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
queueSystem.offService()
```
